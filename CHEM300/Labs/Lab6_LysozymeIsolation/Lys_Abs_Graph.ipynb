```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Lys_Abs.csv')
df.head()
```

```python
X = df.iloc[:, 0]
y = df.iloc[:, 1::2]
```

```python
for i in y.columns:
    plt.plot(X, y[i])

plt.xlabel("Time (s)")
plt.ylabel("Absorbance (A)")
plt.title("Lysozyme Absorbance at 450 nm Over Time")
plt.legend(["Blank", "FT1", "FT2", "E2", "E1", "E3", "Egg Soln"], loc='center left', bbox_to_anchor=(1.05, 0.5))
```
